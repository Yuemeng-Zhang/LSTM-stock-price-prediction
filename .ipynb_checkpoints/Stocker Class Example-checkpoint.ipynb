# Introduction

In this notebook, we will go over the methods in downloader and stocker. And we will use a Xgboost to do prediction as a baseline model. 

We first import the two scripts.

```python
from Downloader import downloader
from Stocker import stocker
```

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

## Downloader

For downloader, we will use it to fetch data from the website, in this example, we use AAPL.

```python
down_a = downloader("data") #the 'data' is a directory where we store the downloaaded file.
```

```python
down_a.pull_stock(['AAPL','^GSPC'],True,start_date = "2009-05-08", end_date="2020-05-08",is_save=True)
#we use pull stock to download the file, it will also return a file that includes the adjusted price
#note also the download process is parllel
```

```python
#now we define a function to read the data 
#in this notebook, we will only use adjusted price, date, and volume, so we will only import them
def adjust_data(stock):
    '''we define this function to specify the columns we will use in the stocker class'''
    filepath = 'data//'+ stock+"_price.csv"
    data = pd.read_csv(filepath)
    data = data[['date','adjclose','volume']]
    data.columns=['Date',stock,'Volume']
    data['Date'] = pd.to_datetime(data['Date'])
    return(data)
```

```python
data = adjust_data('AAPL')
data.head()
#so, this data file is all we want in this project. we will start using it with stocker class now.
```

## Stocker

stocker is a class which has implementations on stock dataframe just as above.
we will go over the methods one by one here.

```python
stocker = stocker(seed=9001) #this is to specify random seed
stocker.import_data(data) #it takes a dataframe as import
stocker.plot_self() #we can plot it use plot_self()
```

Then, we want to generate some theortical indicators, the stocker class includes the following implementation:

1. generate_technical_indicators
2. generate_momentum
3. generate_fft

we will go over them one by one

### generate_technical_indicators

```python
tech_df = stocker.generate_technical_indicators(plot=True, last_days=400)
#last days are the number of days you want to include in the plot
```

```python
tech_df.head()
#a head look at tech_df, it is indexed by Date, which is favorable when we want to join tables.
#some variables are NaN because the definition. For example, you don't have a moving average for the first six days when the window is 7
```

### generate_momentum

```python
# momentum, the inputs are pretty much the same
momentum_df = stocker.generate_momentum(wind_size=30, plot=True, last_days=500)
```

```python
momentum_df[35:40]
```

### generate_fft

```python
#lastly let's go over fft
fft_df = stocker.generate_fft(num_components=[3,5,10,40,60,100],plot=True)
```

```python
fft_df.head()
```

## Split data by rolling timewindow 

Our last feature is to split the dataset by rolling window.

There are two ways to split data, randomly or split sequentialy by a time. These two methods are all supported by stocker class.

```python
#we show an example using random split
#here the random seed is determined when define a stocker class, 
#but we can also use  stocker.edit_seed(num) to change the seed
(X_train,y_train),(X_test,y_test) = stocker.data_split_window(split_ratio=0.8, wind_size=255, 
                                                                      plot=True, random=True)
```

```python
#we will also show an example using sequential split
(X_train,y_train),(X_test,y_test) = stocker.data_split_window(split_ratio=0.8, wind_size=255, 
                                                                      plot=True, random=False)
```

```python
#In this way, we generate train data by a time. let's see the example
X_train.head()
#Note here! Our data starts counting from 2010-05-13, this is because time window=255 is a total year
#and, the previous stock prices all became features.
```

## Join all features

Now, we have these features we want to include:
1. X_train: they are features generated by rolling timewindow
2. fft_df: they are features generated by fft
3. momentum_df: they are features generated by momentum
4. tech_df: they are features generated by theoritcal indicators

We can easily join them all together, or we can join them for any part we want just by doing:

```python
X_train_fin = X_train.join([fft_df,momentum_df,tech_df])
print(X_train_fin.shape)
X_train_fin.head()
```

So, X_train_fin is the ultimate feature we want, and we will do the same thing on X_test just to make sure the feature space is the same.

```python
X_test_fin = X_test.join([fft_df,momentum_df,tech_df])
print(X_test_fin.shape)
X_test_fin.head()
```

## XGboost

At the end, we will train a XGboost as a baseline model.

```python
import xgboost as xgb

regressor = xgb.XGBRegressor(gamma=0.0,objective="reg:squarederror",n_estimators=150,base_score=0.7, 
                             eval_metric='rmse',
                             learning_rate=0.05)

xgbModel = regressor.fit(X_train_fin, y_train.values, \
                         eval_set =  [ (X_train_fin,y_train.values),(X_test_fin,y_test.values) ], \
                         verbose=False)

eval_result = regressor.evals_result()
training_rounds = range(len(eval_result['validation_0']['rmse']))


plt.scatter(x=training_rounds,y=eval_result['validation_0']['rmse'],label='Training Error')
plt.scatter(x=training_rounds,y=eval_result['validation_1']['rmse'],label='Validation Error')
plt.xlabel('Iterations')
plt.ylabel('RMSE')
plt.title('Training Vs Validation Error')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(50,50))
plt.xticks(rotation='vertical')
plt.bar([i for i in range(len(xgbModel.feature_importances_))], xgbModel.feature_importances_.tolist(), tick_label=X_test_fin.columns)
plt.title('Figure 6: Feature importance of the technical indicators.')
plt.show()
```

```python
import_dict = dict(zip(X_test_fin.columns, xgbModel.feature_importances_.tolist()))
import operator
sorted(import_dict.items(), key=operator.itemgetter(1), reverse=True)[0:20]
```

```python
yhat = regressor.predict(X_test_fin)
```

```python
stocker.measurement(yhat)
```

The Above result shows our model is not good at predicting long term trend (of course that is true).
