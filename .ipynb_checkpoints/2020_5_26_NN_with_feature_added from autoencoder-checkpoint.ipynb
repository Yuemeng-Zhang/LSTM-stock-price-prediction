```python
import tensorflow as tf
from tensorflow import keras
import kerastuner as kt
from kerastuner.tuners import RandomSearch, Hyperband, BayesianOptimization

from keras.models import Sequential
from keras.initializers import glorot_normal
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D, Activation
from keras.optimizers import RMSprop, Adam, SGD
from keras.preprocessing.image import ImageDataGenerator
from keras.losses import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

RANDOM_SEED = 42
```

```python
sF = pd.read_hdf('AAPL_X.h5')

PC_80 = pd.read_csv('PC_80.csv',index_col=0)
PC_90 = pd.read_csv('PC_90.csv',index_col=0)

#dataset with PC_80/PC_90 added
sF_80=sF.join(PC_80)
sF_90=sF.join(PC_90)
```

```python
sF
```

```python
class NN_model:
    def __init__(self,RANDOM_SEED):
        self.RANDOM_SEED=RANDOM_SEED
        
    def data_preparation(self,data,test_size=0.2):
        numVal = int(data.shape[0]*0.9)
        X, y  = data.iloc[:numVal, 1:], data.iloc[:numVal, 0]
        X_val, y_val = data.iloc[numVal:, 1:], data.iloc[numVal:, 0]
        
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=self.RANDOM_SEED)
        
        scaler = MinMaxScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        X_val_scaled = scaler.transform(X_val)

        cols = X_train.columns
        X_train_scaled = pd.DataFrame(np.array(X_train_scaled), columns=cols)
        X_test_scaled = pd.DataFrame(np.array(X_test_scaled), columns=cols)
        X_val_scaled = pd.DataFrame(np.array(X_val_scaled), columns=cols)
        self.X_train_scaled, self.X_test_scaled, self.X_val_scaled, self.y_train, self.y_test, self.y_val=X_train_scaled, X_test_scaled, X_val_scaled, y_train, y_test, y_val
        return X_train_scaled, X_test_scaled, X_val_scaled, y_train, y_test, y_val
    
    def model_generate(self,lr=0.0036116627821472483, rho=0.8073248494741022): #here use result from keras tuner
        model = keras.Sequential()
        model.add(keras.layers.Dense(units = 32,
                                     activation="relu", 
                                     input_shape=[self.X_train_scaled.shape[1]]))
        model.add(keras.layers.Dense(units = 8, activation='relu'))
        model.add(keras.layers.Dense(1))

        model.compile(
            optimizer=keras.optimizers.RMSprop(learning_rate = lr, rho = rho),
            loss = 'mse',
            metrics = ['mse','mae'])
        return model
    
    def fit(self,plot=True):
        model = self.model_generate()
        model.fit(self.X_train_scaled, self.y_train, epochs=200, validation_data=(self.X_test_scaled, self.y_test), verbose=0)
        if plot:
            accuracy_df = pd.DataFrame(model.history.history)
            plt.figure(figsize=(17, 10))
            np.sqrt(accuracy_df[['mae', 'val_mae']]).plot()
            plt.title('Loss & Accuracy Per EPOCH For ' + 'Neural Network Model')
            plt.xlabel('EPOCH')
            plt.ylabel('Accruacy')
            plt.show() 
        return model
        
    def plot_predictions(self,model,symbol = 'AAPL'):
        predictions = model.predict(self.X_val_scaled)
        print('predictions shape:', predictions.shape)
        predictions = pd.DataFrame(np.column_stack([self.y_val, predictions]), index = self.y_val.index, columns=['price', 'pred_price'])

        mse = mean_squared_error(predictions['pred_price'], predictions['price'])
        plt.figure(figsize=(17, 10))
        predictions['price'].plot(color='black', label='Price')
        predictions['pred_price'].plot(color='red', label='Predicted')
        plt.xlabel('Date')
        plt.ylabel('USD')
        plt.title('Neural Network model on ' + symbol + ': MSE = ' + str(round(mse, 2)))
        plt.legend()
        plt.show()
        return mse
```

#### original dataset

```python
NN1=NN_model(RANDOM_SEED)
NN1.data_preparation(sF)
model1=NN1.fit()
```

```python
MSE_original=NN1.plot_predictions(model1)
```
